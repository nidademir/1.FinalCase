# **Eksik Değer Analizi**

```python

# Veri setindeki eksik değerlerin tespiti

df.isnull().sum()
```

'Arrival Delay in Minutes' sütununda 310 adet eksik veri var.

```python
# Eksik değer yüzdelerini hesapla
missing_percent = df.isnull().mean() * 100

# Sadece eksik değeri olan sütunları göster
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

# Görüntüle
print(missing_percent)
```

**Tüm veri setinde %0.3 civarında eksiklik var.**

Bu durumda eksik değerler için uygulayacağımız birkaç yöntem var.

1. Ortalama ile Doldurma (Mean Imputation)

  Eksik değerlerin yerine o sütunun ortalamasını koymak.

  ✅ Avantaj:
  *   Basit ve hızlı
  *   Sayısal sütunlar için yaygın yöntem.

  🚫 Dezavantaj:
  *   Aykırı değerler varsa ortalamayı bozabilir.

2. Medyan ile Doldurma (Median Imputation)

  Eğer sütunda aykırı değerler varsa, medyan daha güvenli bir seçimdir.

  ✅ Aykırı değerlerin etkisini azaltır

3. Silme (Dropna)

  Eğer eksik değerli satırları silmek sonuçları bozmayacaksa silme işlemi yapabiliriz. Ama bu durumda 310 örnek kaybederiz.

4. Regresyonla Tahmin Etme (Advanced)

Eksik değeri, diğer sütunları kullanarak bir regresyon modeliyle tahmin edebiliriz. Ancak bu daha ileri düzey bir yöntemdir ve genelde çok fazla eksik veri olduğunda tercih edilir.

**Sonuç:**

Sonuç olarak eksik değer oranı düşük (%0.3) ve sayısal sütun olduğu için **medyan** ile doldurmak genellikle en sağlam ve risksiz yoldur.

# **Aykırı Değer (Outlier) Analizi**

```python
# Sayısal sütunları seçelim
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# IQR yöntemi ile aykırı değer tespiti
def get_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] < lower) | (df[col] > upper)]

# Aykırı değeri olan sütunlar için görselleştirme
for col in num_cols:
    outliers = get_outliers(df, col)
    if not outliers.empty:
        plt.figure(figsize=(8, 2))
        sns.boxplot(x=outliers[col], color='purple')
        plt.title(f"Aykırı Değerler: {col}")
        plt.xlabel(col)
        plt.tight_layout()
        plt.show()
```
